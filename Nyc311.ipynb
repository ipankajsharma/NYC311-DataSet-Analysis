```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#below code is uesed to print multiple outputs in a single cell 
#: refer cell 3 where I have written mupliple EDA commands

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" 
```

```python
#load the file

load_file = pd.read_csv("nyc311.csv")

#make sure nyc311.csv and jupyter notebook are in the same directory, if not use relative path
```

```python
load_file.shape
```

```python
load_file.head()
```

```python
#complaint type and city

complaintTypeCity = pd.core.frame.DataFrame({"count": 
                                             load_file.groupby(['Complaint Type', 'City']).size()}).reset_index()

complaintTypeCity
```

```python
#Major Compliant Types

majorComplaints = load_file.dropna(subset=['Complaint Type'])
majorComplaints = load_file.groupby("Complaint Type")
sortedComplaintType = majorComplaints.size().sort_values(ascending = False)
sortedComplaintType = sortedComplaintType.to_frame('count').reset_index() #making the data presentable

sortedComplaintType
```

```python
#top 10 complaints

sortedComplaintType.head(10)
```

```python
#Bar Graph for Complaintype and Count

complaintType = majorComplaints.size().sort_values(ascending = False)
plt.figure(figsize = (40,40)) #defining the plot size
complaintType.plot(kind='bar', color='g',width=.3,alpha=0.3)
plt.title("Complaint_Type vs Count")
plt.xlabel("Complaint_Type")
plt.ylabel("Count")

plt.show()
```

```python
#major complaint types

sortedComplaintType = sortedComplaintType.head()
plt.figure(figsize = (5,5))
plt.pie(sortedComplaintType['count'],labels=sortedComplaintType['Complaint Type'],autopct='%1.1f%%', shadow=False)

plt.show()
```

```python
#major complaint types and count (Bar graph)

nyc_top10 = complaintType.sort_values(ascending=False)
nyc_top10 = nyc_top10.head(10)
plt.figure(figsize = (5,5))
nyc_top10.plot(kind='bar',color='b',width=.5,alpha=0.5)

plt.title("Types of Complaint")
plt.xlabel("Complaint_Type")
plt.ylabel("Count")

plt.show()
```
